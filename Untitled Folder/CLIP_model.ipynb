# Importing

```python
from PIL import Image
import requests
from transformers import CLIPProcessor, CLIPModel
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (4, 2)

import torch
import itertools

import torchvision
import numpy as np
```

# Loading Model

```python
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

# Testing

```python
def get_rainbow_list(gradients_per_channel=range(255)):
    color_list = torch.zeros([len(gradients_per_channel)**3, 3])
    index = 0
    for color in itertools.product(gradients_per_channel, repeat=3):
        color_list[index] = torch.tensor(color)
        index += 1
    return color_list

device = 'cuda' if torch.cuda.is_available() else 'cpu'
image = get_rainbow_list(gradients_per_channel=range(0, 255, 40)).type(torch.float).to(device).unsqueeze(1).unsqueeze(1).repeat(1, 10, 10, 1)
print(image.shape)
```

```python
def plot_color_association_bar_chart(color_list, hist_values):
    plt.bar(x=range(len(hist_values)), height=hist_values, color=color_list)
    plt.show()
```

```python
inputs = processor(text=["apple", "blueberry", "banana", "grape", "cherry", "tangerine"], images=[image[sample].permute(2, 0, 1) for sample in range(image.shape[0])], return_tensors="pt", padding=True)
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image  # this is the image-text similarity score

logits_per_text = outputs.logits_per_text.softmax(dim=1)  # this is the image-text similarity score

print(logits_per_text.shape)
```

```python
for word in range(logits_per_text.shape[0]):
    plot_color_association_bar_chart([image[sample, 0, 0].cpu().numpy()/255 for sample in range(image.shape[0])], logits_per_text[word].cpu().detach().numpy())
```

```python
inputs = processor(text=["apple clipart", "blueberry clipart", "banana clipart", "grape clipart", "cherry clipart", "tangerine clipart"], images=[image[sample].permute(2, 0, 1) for sample in range(image.shape[0])], return_tensors="pt", padding=True)
outputs = model(**inputs)
logits_per_image = outputs.logits_per_image  # this is the image-text similarity score

logits_per_text = outputs.logits_per_text.softmax(dim=1)  # this is the image-text similarity score

print(logits_per_text.shape)
```
