```python
def is_grayscale(img_path):
    with Image.open(img_path).convert('RGB') as img:
        w, h = img.size
        for i in range(w):
            for j in range(h):
                r, g, b = img.getpixel((i,j))
                if r != g != b: 
                    return False
    return True
```

```python
urls = []
grayscale = []
data_save_path = os.path.join("data", "scraped_image_grayscale2.csv")

count = 0 # change back to 0
for url in tqdm.tqdm(data["URL"]):
    count += 1
#     print(url)
    try:
        with requests.get(url, timeout=10) as response:
            response.raise_for_status()
            is_gray = is_grayscale(BytesIO(response.content))
            urls.append(url)
            if is_gray:
                grayscale.append(1)
            else:
                grayscale.append(0)

    except:
        pass
        
    if count % 1000 == 0:
        df = pd.DataFrame(np.array([urls, grayscale]).T, columns=["url", "grayscale"])
        df.to_csv(data_save_path)
        
df = pd.DataFrame(np.array([urls, grayscale]), columns=["url", "grayscale"])
df.to_csv(data_save_path)
print(sum(grayscale) / len(grayscale))
df
```
