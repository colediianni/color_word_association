# Importing

```python
from PIL import Image
import requests
from transformers import CLIPProcessor, CLIPModel
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (4, 2)

import torch
import itertools

import torchvision
import numpy as np
```

# Loading Model

```python
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

# Testing

```python
def get_rainbow_list(gradients_per_channel=range(255)):
    color_list = torch.zeros([len(gradients_per_channel)**3, 3])
    index = 0
    for color in itertools.product(gradients_per_channel, repeat=3):
        color_list[index] = torch.tensor(color) / 255
        index += 1
    return color_list

device = 'cuda' if torch.cuda.is_available() else 'cpu'
image = get_rainbow_list(gradients_per_channel=range(0, 255, 40)).type(torch.float).to(device).unsqueeze(1).unsqueeze(1).repeat(1, 10, 10, 1)
print(image.shape)
```

```python
def plot_color_association_bar_chart(color_list, hist_values, title):
    plt.title(title)
    plt.bar(x=range(len(hist_values)), height=hist_values, color=color_list)
    plt.show()
```

```python
def get_color_word_associations(text, colors):
    with torch.no_grad():
        inputs = processor(text=text, images=colors, return_tensors="pt", padding=True)
        outputs = model(**inputs)
        logits_per_image = outputs.logits_per_image  # this is the image-text similarity score
        logits_per_text = outputs.logits_per_text.softmax(dim=1)
        return logits_per_text
```

```python
text = ["the color red", "the color blue"]
cwa = get_color_word_associations(text, [image[sample].permute(2, 0, 1) for sample in range(image.shape[0])])

for word in range(cwa.shape[0]):
    plot_color_association_bar_chart([image[sample, 0, 0].cpu().numpy() for sample in range(image.shape[0])], 
                                    cwa[word].cpu().detach().numpy(),
                                    text[word])
```

```python
text = ["apple", "blueberry", "banana", "grape", "cherry", "tangerine"]
cwa = get_color_word_associations(text, [image[sample].permute(2, 0, 1) for sample in range(image.shape[0])])

for word in range(cwa.shape[0]):
    plot_color_association_bar_chart([image[sample, 0, 0].cpu().numpy() for sample in range(image.shape[0])], 
                                    cwa[word].cpu().detach().numpy(),
                                    text[word])
```
